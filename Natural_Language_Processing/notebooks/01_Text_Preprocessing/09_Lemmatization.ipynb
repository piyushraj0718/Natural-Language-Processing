# Lemmatization

## Definition
Lemmatization is a text normalization technique that reduces a word to its dictionary base form, known as the **lemma**, using vocabulary and morphological analysis. Unlike stemming, lemmatization always produces a valid word. For example, `better` → `good`, `was` → `be`, `running` → `run`.

## Why It Is Needed
- **Linguistic Accuracy:** Produces real, meaningful words instead of truncated stems, preserving semantic integrity.
- **Better Semantic Grouping:** Correctly maps irregular forms — `went` → `go`, `children` → `child` — which stemming cannot handle.
- **Improved NLP Model Quality:** More accurate base forms lead to better word embeddings and downstream task performance.

## Real-World Applications
- Question answering systems where word meaning must be preserved
- Chatbot intent recognition preprocessing
- Sentiment analysis where word polarity depends on correct base forms
- Legal and medical document analysis requiring linguistic precision
- Machine translation preprocessing pipelines

## Important Points
- **Lemma:** The canonical, dictionary form of a word. e.g., `running` → `run`, `geese` → `goose`.
- **POS Dependency:** Lemmatization requires the Part-of-Speech (POS) tag to resolve ambiguity.
  - `meeting` as a **verb** → `meet`
  - `meeting` as a **noun** → `meeting`
- **WordNet Lemmatizer (NLTK):** Uses the WordNet lexical database for lemmatization.
- **spaCy Lemmatizer:** Integrated into spaCy's pipeline; automatically uses POS tags for accurate lemmas.
- **Stemming vs. Lemmatization:**

| Feature | Stemming | Lemmatization |
|---|---|---|
| Output | May not be a real word | Always a valid dictionary word |
| Speed | Faster (rule-based) | Slower (uses lexicon + POS) |
| Accuracy | Lower | Higher |
| Handles Irregular Forms | No | Yes |

```python
import nltk
import spacy

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from collections import Counter

# Download required resources
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')

# Initialize lemmatizers

lemmatizer = WordNetLemmatizer()
nlp = spacy.load("en_core_web_sm")

# Sample corpus from different domains
corpus = {
    "news": "The economic downturn is affecting global markets. Investors are reconsidering their portfolios.",
    "wikipedia": "The mitochondrion is often referred to as the powerhouse of the cell.",
    "science_paper": "In recent studies, convolutional neural networks have demonstrated state-of-the-art performance in image classification tasks."
}

# Tokenize corpus
tokenized_corpus = {key: nltk.word_tokenize(value) for key, value in corpus.items()}
```

```python
def apply_lemmatization(text_tokens, lemmatizer, pos=wordnet.VERB):
    return [lemmatizer.lemmatize(word, pos) for word in text_tokens]

for domain, tokens in tokenized_corpus.items():
    print(f"\n**{domain.upper()} DOMAIN**")
    print("NLTK Lemmatization:", apply_lemmatization(tokens, lemmatizer))
```

```python
def spacy_lemmatization(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc]

for domain, text in corpus.items():
    print(f"\n**{domain.upper()} DOMAIN**")
    print("SpaCy Lemmatization:", spacy_lemmatization(text))
```
