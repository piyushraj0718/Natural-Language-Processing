# Frequency Distribution

## Definition
Frequency distribution is a statistical summary that maps each unique token in a corpus to the number of times it appears. It provides a complete picture of how often each word, bigram, or n-gram occurs, enabling analysis of vocabulary richness, dominant themes, and linguistic patterns in text data.

## Why It Is Needed
- **Vocabulary Insight:** Reveals which words are most and least informative in a corpus.
- **Feature Selection:** Helps filter very rare tokens (noise) and very frequent tokens (stop words) before building models.
- **Exploratory Text Analysis:** Gives a quick statistical snapshot of what a corpus is about before applying complex models.

## Real-World Applications
- Generating word clouds for document visualization
- Identifying trending keywords in social media data
- Analysing n-gram frequency for language model training
- Detecting the vocabulary distribution of different text domains
- Monitoring keyword frequency changes over time in news corpora

## Important Points
- **NLTK `FreqDist`:** A built-in class in NLTK that computes and visualizes frequency distributions.
- **N-gram Frequency:** Extending frequency distribution to bigrams or trigrams captures phrase-level patterns.
  - Bigram: `("natural", "language")`, Trigram: `("natural", "language", "processing")`
- **Cumulative Frequency Plot:** Shows what percentage of the total corpus is covered by the top-N most frequent words.
- **Hapax Legomena:** Words that appear exactly once in a corpus — often indicate noise or rare terminology.
- Zipf's Law governs the shape of any natural language frequency distribution.

````python
# words feuency distribution

from collections import Counter
import nltk
from nltk.tokenize import word_tokenize

nltk.download("punkt_tab")

def word_frequency(text):
    words = word_tokenize(text.lower())
    return Counter(words)

text = "NlP is amazing , NLP makes machine understands human language "

print(word_frequency(text))
````

# TF-IDF - Term Frequency-Inverse Document Frequency

* It is a statistical measure used in natural language processing (NLP) and search engines to evaluate how important a word is to a specific document within a larger collection (corpus)

````python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print("Feature Names:", vectorizer.get_feature_names_out())
print("TF-IDF Representation:")
print(X.toarray())
````

````python
import numpy as np

feature_array = np.array(vectorizer.get_feature_names_out())
importance = np.argsort(X.toarray()).flatten()[::-1]

keywords = feature_array[importance[:5]]
print("Top Keywords:", keywords)
````
