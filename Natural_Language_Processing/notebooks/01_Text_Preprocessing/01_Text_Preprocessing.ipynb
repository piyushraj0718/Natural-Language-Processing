# Text Preprocessing

## Definition
Text preprocessing is the process of cleaning and transforming raw text into a structured format suitable for NLP tasks. Raw text from real-world sources is noisy, inconsistent, and unstructured. Preprocessing converts it into a clean, machine-readable form. It is the mandatory first step in any NLP pipeline.

## Why It Is Needed
- **Noise Removal:** Real-world text contains HTML tags, special characters, emojis, and irrelevant symbols that can mislead models.
- **Standardization:** Different spellings, casing, and formats for the same word must be unified before analysis.
- **Model Performance:** Clean, well-structured input directly improves the accuracy and efficiency of downstream NLP models.

## Real-World Applications
- Cleaning user reviews before sentiment analysis
- Preprocessing tweets for social media monitoring
- Sanitizing customer support tickets for topic classification
- Preparing news articles for summarization systems
- Cleaning medical records for clinical NLP models

## Important Points
- **Pipeline Order:** Lowercasing → Noise Removal → Tokenization → Stop-word Removal → Stemming/Lemmatization
- **Lowercasing:** Ensures `Apple` and `apple` are treated as the same token.
- **Noise Removal:** Strips HTML, URLs, punctuation, numbers, and extra whitespace.
- **Stop-word Removal:** Removes high-frequency, low-information words like *the*, *is*, *and*.
- Preprocessing steps vary by task — e.g., punctuation matters for NER but not for topic modelling.

# Stop word removal 

* it is the process of removing common words that do not carry significant meaning in a sentence. These words are often referred to as "stop words" and include words like "the," "is," "in," "and," etc. Removing stop words helps reduce noise in the text and improves the efficiency of NLP models.

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))
print(stop_words)
print(len(stop_words))
```

# Lowercasing

* it is the process of converting all characters in a text to lowercase. Lowercasing is important in NLP because it ensures that words with different cases (e.g., "Apple" and "apple") are treated as the same token, reducing redundancy and improving model performance.

# Remove Punctutation

* it is the process of eliminating punctuation marks from text. Punctuation can introduce noise and inconsistencies in NLP tasks, so removing it helps create cleaner and more uniform data for analysis.

```python
import string 
# sample 

text = "Hello, World! This is a sample text with punctuation marks. Let's see how we can remove them."


```

```python
print(string.punctuation)
print(len(string.punctuation))
```

```python
translator = str.maketrans("", "", string.punctuation)
processed = text.translate(translator)

print(processed)
```

# Stemming

* it is the process of reducing words to their base or root form by removing suffixes and prefixes. Stemming helps group similar words together, improving the efficiency of NLP models. For example, "running," "runner," and "ran" can all be reduced to the stem "run."

```python
from nltk.stem import PorterStemmer

# sample words

words = ["running", "runner", "ran", "easily", "fairly"]

stemer = PorterStemmer()
stemed = [stemer.stem(word) for word in words]
print(stemed)
```

# POS Tagging

* it is the process of assigning parts of speech (POS) to each word in a sentence, such as nouns, verbs, adjectives, etc. POS tagging provides syntactic information about the words, which can be useful for various NLP tasks like parsing, named entity recognition, and sentiment analysis.

```python
from nltk import pos_tag

text = "I am trying to give my best"

tokens= word_tokenize(text)
print((tokens))
print(len(tokens))
```

```python
tagged = pos_tag(tokens)
print(tagged)
```

# Lemmatization

* it is the process of reducing words to their base or dictionary form, known as the lemma. Unlike stemming, lemmatization considers the context and meaning of the word, ensuring that the resulting lemma is a valid word. For example, "better" would be lemmatized to "good," and "running" would be lemmatized to "run." Lemmatization helps improve the accuracy of NLP models by providing more meaningful representations of words.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
words = "runner burnt happily flying"

doc = nlp(words)

for token in doc:
    print(token.text, token.lemma_)
```

```python
for token in doc:
    print(f"{token.text :<15} {token.pos_ :<10} {token.lemma_}")
```

# N-Grams

* it is a contiguous sequence of n items (words, characters, or tokens) from a given text. N-grams are used to capture context and relationships between words in NLP tasks. For example, in the sentence "I love natural language processing," the bigrams (2-grams) would be "I love," "love natural," "natural language," and "language processing." N-grams are useful for tasks like text classification, language modeling, and machine translation.

```python
from nltk.util import ngrams
```

```python
sample_text= " My name is Piyush Raj , and i am  not a procastinator "
sample_text = sample_text.lower()
```

```python
tokens = word_tokenize(sample_text)
tokens
```

```python
tokens = [word for word in tokens if word not in stop_words]
```

```python
bigrams = list(ngrams(tokens , 2))

print(bigrams)
print(len(bigrams))
```
