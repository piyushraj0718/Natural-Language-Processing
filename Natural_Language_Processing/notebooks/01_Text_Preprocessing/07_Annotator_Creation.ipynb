# Annotator Creation

## Definition
An annotator in NLP is a component or pipeline stage that adds linguistic metadata — such as part-of-speech tags, named entities, dependency labels, or custom labels — to raw text. Annotator creation involves designing and building these labeling components, either using rule-based systems, pre-trained models, or custom-trained classifiers.

## Why It Is Needed
- **Structured Labeling:** Raw text has no structure; annotators assign meaningful labels that downstream models can learn from.
- **Custom Domain Adaptation:** Pre-built annotators may not handle domain-specific language (medical, legal, financial) — custom annotators fill this gap.
- **Pipeline Automation:** Annotators enable automated labeling at scale, reducing manual effort in data preparation.

## Real-World Applications
- Creating custom Named Entity Recognition (NER) annotators for medical records
- Building POS taggers for low-resource languages
- Annotating sentiment polarity in customer feedback datasets
- Labelling intent and entities in conversational AI training data
- Creating annotation pipelines for document classification tasks

## Important Points
- **Rule-Based Annotators:** Use regex or lexicons to label text — fast but brittle.
- **Model-Based Annotators:** Use trained ML/DL models (e.g., CRF, BiLSTM, Transformers) for labeling.
- **spaCy Pipeline Components:** Custom annotators can be added as pipeline components using `@Language.component` or `@Language.factory`.
- **NLTK Taggers:** NLTK provides `DefaultTagger`, `RegexpTagger`, `UnigramTagger`, and `BigramTagger` as building blocks.
- **Annotation Schema:** Defining consistent label sets (e.g., BIO tagging for NER) is crucial before building annotators.
  - **BIO Format:** `B-` (Beginning), `I-` (Inside), `O` (Outside)

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
pos_sentence = "NlP is Amazing"
ner_sentence = "Google was invented by Larry Page and Sergey Brin"
```

```python
# Pos Tagging 

pos_doc = nlp(pos_sentence)
pos_tags = [(token.text , token.pos_) for token in pos_doc]

print(pos_tags)
```

```python
# ner

ner_doc = nlp(ner_sentence)
ner_tags = [(ent.text , ent.label_) for ent in ner_doc.ents]

print(ner_tags)
```
