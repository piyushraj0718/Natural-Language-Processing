# String Tokenization

## Definition
Tokenization is the process of splitting a string of text into smaller units called tokens. Tokens can be words, subwords, characters, or sentences depending on the granularity required. It is the most fundamental step in converting raw text into a form that NLP models can process.

## Why It Is Needed
- **Unit of Analysis:** NLP models operate on tokens, not raw strings — tokenization defines the basic unit of processing.
- **Vocabulary Building:** Every unique token becomes an entry in the model's vocabulary, enabling numerical representation.
- **Boundary Detection:** Correctly splitting sentences and words ensures that meaning is preserved during downstream processing.

## Real-World Applications
- Splitting sentences into words for bag-of-words models
- Subword tokenization in transformer models like BERT and GPT
- Sentence tokenization for document summarization
- Character-level tokenization for spell correction systems
- Tweet tokenization handling hashtags, mentions, and emojis

## Important Points
- **Word Tokenization:** Splits text on whitespace and punctuation. e.g., `"I am happy"` → `["I", "am", "happy"]`.
- **Sentence Tokenization:** Splits a paragraph into individual sentences.
- **Subword Tokenization:** Handles unknown words by splitting into known subword units.
  - **BPE (Byte Pair Encoding):** Used in GPT models.
  - **WordPiece:** Used in BERT (`playing` → `play` + `##ing`).
  - **SentencePiece:** Language-agnostic tokenization used in multilingual models.
- **Character Tokenization:** Each character is a token — useful for morphologically rich languages.
- Whitespace-based tokenization is the simplest but fails on languages like Chinese and Japanese.

## Key Takeaways
- Tokenization converts raw text into discrete units for model processing.
- The choice of tokenizer type (word, subword, char) affects vocabulary size and model performance.

# Text Summarization 

```python
import spacy
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Input text
text = """Natural Language Processing (NLP) is a field of AI that enables machines to understand human language.
It is used in chatbots, sentiment analysis, and language translation.
One of the key challenges in NLP is understanding context and ambiguity in sentences.
Deep learning models like transformers have significantly improved NLP applications."""

# Process text with spaCy
doc = nlp(text)

# Tokenize words and compute frequency
words = [word.lower() for word in word_tokenize(text) if word.isalnum()]
filtered_freq = Counter(words)

# Compute sentence scores
sentence_scores = {}
for sent in doc.sents:
    for word in word_tokenize(sent.text.lower()):
        if word in filtered_freq:
            sentence_scores[sent.text] = sentence_scores.get(sent.text, 0) + filtered_freq[word]

# Sort sentences by score
sorted_sentences = sorted(sentence_scores, key=sentence_scores.get, reverse=True)

# Select top-ranked sentences for summary
summary = " ".join(sorted_sentences[:2])

# Print summary
print("Summary:", summary)
```
