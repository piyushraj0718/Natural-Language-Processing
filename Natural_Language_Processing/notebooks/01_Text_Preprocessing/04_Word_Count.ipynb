# Word Count

## Definition
Word count refers to the process of counting the total number of words or the frequency of each individual word in a text corpus. It is one of the simplest yet most informative analyses in NLP, forming the foundation for statistical text representations like Bag of Words (BoW) and TF-IDF.

## Why It Is Needed
- **Corpus Statistics:** Understanding word frequency distribution gives insight into the vocabulary and dominant themes of a corpus.
- **Feature Engineering:** Word counts serve as raw features in text classification, clustering, and retrieval models.
- **Vocabulary Pruning:** Identifying very rare or very common words helps decide which tokens to include or exclude from the vocabulary.

## Real-World Applications
- Identifying the most frequently used terms in customer reviews
- Measuring keyword density for SEO analysis
- Building Bag-of-Words representations for text classifiers
- Analysing word usage patterns across different authors or time periods
- Detecting dominant topics in a news corpus

## Important Points
- **Total Word Count:** Sum of all tokens in the corpus.
- **Unique Word Count (Vocabulary Size):** Number of distinct tokens.
- **Term Frequency (TF):** `TF(t) = count of term t in document / total terms in document`.
- **Zipf's Law:** In natural language, word frequency is inversely proportional to its rank — the most common word appears roughly twice as often as the second most common word.
- Stop words like *the*, *a*, *is* dominate raw word counts and often need to be removed.

```python
#regex - handle punctuation and count words 


import re 
def count_words(text):
    words = re.findall(r'\b\w+\b', text)
    return len(words)

text = "hello world ! , welcome to NLP"
print("Word Count :", count_words(text))
```

```python
# NLTK - Python 

import nltk

from nltk import word_tokenize 

def count_words_nltk(text):

    words = word_tokenize(text)
    return len(words)

text = "Let's explore NLP with Python!"
print(word_tokenize(text))
print("Word Count:", count_words_nltk(text))
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def count_words_spacy(text):
    doc = nlp(text)
    words = [token.text for token in doc if not token.is_punct]
    return len(words)

text = "NLP is amazing, isn't it?"
doc = nlp(text)
print([token.text for token in doc])
print("Word Count:", count_words_spacy(text))
```

```python
from collections import Counter

def word_frequency(text):
    words = text.lower().split()
    return Counter(words)

text = "NLP is fun. NLP is powerful. NLP is the future."
print("Word Frequencies:", word_frequency(text))
```

# NLP Task 1 - Short vs Long sentences 

```python
from nltk.tokenize import sent_tokenize, word_tokenize

def categorize_sentences(text):
    sentences = sent_tokenize(text)
    categorized = {"short":[] , "long":[]}

    for sentence in sentences:
        word_count = len(word_tokenize(sentence))

        if word_count <5:
            categorized["short"].append(sentence)
        else:
            categorized["long"].append(sentence)

    return categorized

text = "Hello! NLP is fun. Let's learn NLP together. It helps in many fields."
categories = categorize_sentences(text)
print("Short Sentences:", categories["short"])
print("Long Sentences:", categories["long"])
```

# rule based sentiment analysis 

```python
import time

positive_words = {"happy", "good", "great", "excellent", "love", "amazing"}
negative_words = {"sad", "bad", "terrible", "hate", "awful", "worst"}

def sentiment_analysis(text):
    start_time = time.time()
    words = text.lower().split()
    pos_count = sum(1 for word in words if word in positive_words)
    neg_count = sum(1 for word in words if word in negative_words)

    sentiment = "Positive Sentiment" if pos_count > neg_count else "Negative Sentiment" if neg_count > pos_count else "Neutral Sentiment"

    end_time = time.time()
    execution_time = end_time - start_time

    return sentiment, execution_time

text = """
I'm happy you're the President, and thank you for bringing me home.
I have never been so proud to be an American citizen. Thank you, Mr. President.
"""
sentiment, exec_time = sentiment_analysis(text)
print("Sentiment:", sentiment)
print("Execution Time: {:.6f} seconds".format(exec_time))
```

```python
import time

spam_words = {"free", "win", "winner", "prize", "money", "cash", "offer", "click", "buy", "urgent"}
not_spam_words = {"hello", "meeting", "project", "thanks", "family", "friend", "schedule", "regards"}

def spam_detection(text):
    start_time = time.time()

    words = text.lower().split()
    
    spam_count = sum(1 for word in words if word in spam_words)
    not_spam_count = sum(1 for word in words if word in not_spam_words)

    result = "Spam Message" if spam_count > not_spam_count else "Not Spam Message"

    end_time = time.time()
    execution_time = end_time - start_time

    return result, execution_time


text = """
Congratulations! You are a winner.
Click here to claim your free cash prize now.
"""

result, exec_time = spam_detection(text)

print("Classification:", result)
print("Execution Time: {:.6f} seconds".format(exec_time))
```
