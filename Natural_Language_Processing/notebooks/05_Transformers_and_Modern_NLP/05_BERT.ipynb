# BERT (Bidirectional Encoder Representations from Transformers)

## Definition
BERT is a pre-trained Transformer-based language model developed by Google (Devlin et al., 2018). It uses only the Encoder stack of the Transformer and is pre-trained on massive text corpora using two self-supervised tasks: Masked Language Modeling (MLM) and Next Sentence Prediction (NSP). BERT produces deep bidirectional contextual word representations, making it highly effective for a wide range of NLP tasks through fine-tuning.

## Why It Is Needed
- **Bidirectional Context:** Previous models like ELMo and GPT were unidirectional — BERT reads the entire sentence simultaneously in both directions, producing richer contextual representations.
- **Transfer Learning for NLP:** BERT's pre-training on massive corpora captures general language knowledge, which can be fine-tuned for downstream tasks with minimal labeled data.
- **State-of-the-Art Results:** BERT achieved new benchmarks across 11 NLP tasks upon release, fundamentally changing the field.

## Real-World Applications
- Search engine query understanding (used in Google Search since 2019)
- Question Answering systems (SQuAD benchmark)
- Named Entity Recognition (NER)
- Text classification and sentiment analysis
- Natural Language Inference (NLI)

## Important Points
- **Architecture:** Encoder-only Transformer.
  - `BERT-Base`: 12 layers, 768 hidden size, 12 attention heads — 110M parameters.
  - `BERT-Large`: 24 layers, 1024 hidden size, 16 attention heads — 340M parameters.
- **Pre-training Tasks:**
  - **MLM (Masked Language Modeling):** 15% of input tokens are randomly masked — the model predicts them using bidirectional context.
  - **NSP (Next Sentence Prediction):** Given two sentences A and B, predict whether B follows A — helps learn sentence relationships.
- **Special Tokens:**
  - `[CLS]` — Prepended to every input; its final representation is used for classification tasks.
  - `[SEP]` — Separates two sentences in the input.
  - `[MASK]` — Replaces masked tokens during MLM pre-training.
- **Fine-Tuning:** Add a task-specific output layer on top of `[CLS]` or token representations and fine-tune all weights end-to-end.
- **WordPiece Tokenization:** BERT uses WordPiece tokenization with a 30,000-token vocabulary.
- **Notable Variants:** RoBERTa (improved pre-training), DistilBERT (compressed), ALBERT (parameter sharing), BioBERT (biomedical).

```python
from transformers import TFBertForMaskedLM, BertTokenizer
import tensorflow as tf

# Load tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertForMaskedLM.from_pretrained('bert-base-uncased')

# Input sentence with proper mask token
input_text = f"The cat sat on the {tokenizer.mask_token} and looked out the window."

# Tokenize
inputs = tokenizer(input_text, return_tensors="tf")

# Run model
outputs = model(inputs)
logits = outputs.logits

# Find index of [MASK] token
mask_token_index = tf.where(inputs["input_ids"] == tokenizer.mask_token_id)[0][1].numpy()

# Predict token at [MASK]
predicted_token_id = tf.argmax(logits[0, mask_token_index]).numpy()
predicted_token = tokenizer.decode([predicted_token_id])

top_k = 5
top_k_ids = tf.math.top_k(logits[0, mask_token_index], k=top_k).indices.numpy()
top_k_tokens = tokenizer.convert_ids_to_tokens(top_k_ids)

print("Top predictions:")
for i, token in enumerate(top_k_tokens, 1):
    print(f"{i}: {token}")
```

```python
!pip install -q tensorflow transformers datasets

import tensorflow as tf
from transformers import DistilBertTokenizer, TFDistilBertForSequenceClassification, create_optimizer
from datasets import load_dataset

# 1. Load small IMDB subset
dataset = load_dataset("imdb")
small_train = dataset["train"].select(range(500))
small_test = dataset["test"].select(range(100))

# 2. Load tokenizer & model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

# 3. Tokenize the text
def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=128)

small_train = small_train.map(tokenize, batched=True)
small_test = small_test.map(tokenize, batched=True)

small_train.set_format(type="tensorflow", columns=["input_ids", "attention_mask", "label"])
small_test.set_format(type="tensorflow", columns=["input_ids", "attention_mask", "label"])

train_tf_dataset = small_train.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols="label",
    shuffle=True,
    batch_size=16
)

test_tf_dataset = small_test.to_tf_dataset(
    columns=["input_ids", "attention_mask"],
    label_cols="label",
    shuffle=False,
    batch_size=16
)

# 4. Hugging Face create_optimizer (FIX)
num_train_steps = len(train_tf_dataset) * 1  # epochs = 1
optimizer, lr_schedule = create_optimizer(
    init_lr=5e-5,
    num_train_steps=num_train_steps,
    num_warmup_steps=0
)

model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=["accuracy"]
)

# 5. Train model (1 epoch)
model.fit(train_tf_dataset, validation_data=test_tf_dataset, epochs=1)

# 6. Predict on custom input
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors="tf", truncation=True, padding="max_length", max_length=128)
    logits = model(inputs).logits
    probs = tf.nn.softmax(logits, axis=1).numpy()[0]
    label = "Positive 😊" if probs[1] > probs[0] else "Negative 😞"
    confidence = round(probs.max(), 2)
    return label, confidence

# 7. Example predictions
examples = [
    "This movie was amazing! I loved every moment.",
    "Worst movie ever. Total waste of time.",
    "It was okay, not great but not terrible either."
]

for example in examples:
    sentiment, conf = predict_sentiment(example)
    print(f"\nInput: {example}\nPrediction: {sentiment} (Confidence: {conf})")
```
