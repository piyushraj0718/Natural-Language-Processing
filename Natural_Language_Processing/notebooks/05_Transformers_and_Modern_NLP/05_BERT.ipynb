# BERT (Bidirectional Encoder Representations from Transformers)

## Definition
BERT is a pre-trained Transformer-based language model developed by Google (Devlin et al., 2018). It uses only the Encoder stack of the Transformer and is pre-trained on massive text corpora using two self-supervised tasks: Masked Language Modeling (MLM) and Next Sentence Prediction (NSP). BERT produces deep bidirectional contextual word representations, making it highly effective for a wide range of NLP tasks through fine-tuning.

## Why It Is Needed
- **Bidirectional Context:** Previous models like ELMo and GPT were unidirectional — BERT reads the entire sentence simultaneously in both directions, producing richer contextual representations.
- **Transfer Learning for NLP:** BERT's pre-training on massive corpora captures general language knowledge, which can be fine-tuned for downstream tasks with minimal labeled data.
- **State-of-the-Art Results:** BERT achieved new benchmarks across 11 NLP tasks upon release, fundamentally changing the field.

## Real-World Applications
- Search engine query understanding (used in Google Search since 2019)
- Question Answering systems (SQuAD benchmark)
- Named Entity Recognition (NER)
- Text classification and sentiment analysis
- Natural Language Inference (NLI)

## Important Points
- **Architecture:** Encoder-only Transformer.
  - `BERT-Base`: 12 layers, 768 hidden size, 12 attention heads — 110M parameters.
  - `BERT-Large`: 24 layers, 1024 hidden size, 16 attention heads — 340M parameters.
- **Pre-training Tasks:**
  - **MLM (Masked Language Modeling):** 15% of input tokens are randomly masked — the model predicts them using bidirectional context.
  - **NSP (Next Sentence Prediction):** Given two sentences A and B, predict whether B follows A — helps learn sentence relationships.
- **Special Tokens:**
  - `[CLS]` — Prepended to every input; its final representation is used for classification tasks.
  - `[SEP]` — Separates two sentences in the input.
  - `[MASK]` — Replaces masked tokens during MLM pre-training.
- **Fine-Tuning:** Add a task-specific output layer on top of `[CLS]` or token representations and fine-tune all weights end-to-end.
- **WordPiece Tokenization:** BERT uses WordPiece tokenization with a 30,000-token vocabulary.
- **Notable Variants:** RoBERTa (improved pre-training), DistilBERT (compressed), ALBERT (parameter sharing), BioBERT (biomedical).

## Key Takeaways
- BERT is a bidirectional Transformer encoder pre-trained on MLM and NSP tasks.
- The `[CLS]` token's representation is used as the sentence-level feature for classification.
- WordPiece tokenization splits unknown words into known subword units.
- Fine-tuning BERT requires minimal labeled data and achieves strong performance.
- BERT and its variants (RoBERTa, DistilBERT) remain among the most widely deployed NLP models.
