# Self-Attention

## Definition
Self-Attention (also called Intra-Attention) is a mechanism where each token in a sequence attends to all other tokens in the same sequence to compute a contextualized representation. Unlike standard attention (which relates encoder to decoder states), self-attention relates each position in a sequence to every other position within the same sequence, enabling the model to capture long-range dependencies without recurrence.

## Why It Is Needed
- **Parallel Processing:** Unlike RNNs, self-attention processes all tokens simultaneously — enabling massive parallelization on GPUs.
- **Long-Range Dependencies:** Captures relationships between any two tokens regardless of their distance in the sequence — no vanishing gradient over distance.
- **Contextual Representations:** Each token's representation is updated based on its relationship to all other tokens, producing deeply contextual embeddings.

## Real-World Applications
- Core computation in all Transformer-based models (BERT, GPT, T5, etc.)
- Coreference resolution (e.g., linking `it` to `the cat` in a sentence)
- Dependency parsing — identifying syntactic relationships between words
- Semantic role labeling
- Any NLP task requiring understanding of word relationships within a sentence

## Important Points
- **Query, Key, Value (Q, K, V):** The three projections at the heart of self-attention.
  - **Query (Q):** What the current token is looking for.
  - **Key (K):** What each token offers or represents.
  - **Value (V):** The actual content to be aggregated.
- **Scaled Dot-Product Attention Formula:**
  - `Attention(Q, K, V) = softmax( Q·Kᵀ / √dₖ ) · V`
  - `dₖ` = dimension of key vectors (scaling factor to prevent vanishing gradients from large dot products).
- **Attention Matrix:** An N×N matrix (N = sequence length) where each cell `[i,j]` holds the attention weight of token `i` attending to token `j`.
- **Why Scale by √dₖ?** Without scaling, dot products grow large in magnitude with high-dimensional keys, pushing softmax into saturation regions with near-zero gradients.
- Q, K, V are all linear projections of the same input sequence — hence `self`-attention.
