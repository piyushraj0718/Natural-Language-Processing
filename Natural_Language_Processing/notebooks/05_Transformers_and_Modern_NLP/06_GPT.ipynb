# GPT (Generative Pre-trained Transformer)

## Definition
GPT (Generative Pre-trained Transformer) is a family of large language models developed by OpenAI that use only the Decoder stack of the Transformer architecture. GPT is pre-trained on massive text corpora using Causal Language Modeling (CLM) — predicting the next token given all previous tokens. It is an autoregressive model, making it naturally suited for text generation, completion, and instruction-following tasks.

## Why It Is Needed
- **Powerful Text Generation:** Autoregressive training on internet-scale data enables GPT to generate fluent, coherent, and contextually appropriate text.
- **Few-Shot Learning:** Large GPT models (GPT-3+) can perform new tasks from just a few examples in the prompt — without any gradient updates (in-context learning).
- **General-Purpose NLP:** A single GPT model can handle diverse tasks — translation, summarization, coding, QA — through prompting alone.

## Real-World Applications
- Conversational AI and chatbots (ChatGPT)
- Code generation and completion (GitHub Copilot)
- Content creation, copywriting, and creative writing
- Document summarization and question answering
- Instruction-following AI assistants

## Important Points
- **Architecture:** Decoder-only Transformer with Masked (Causal) Self-Attention.
- **Pre-training Objective — Causal Language Modeling (CLM):**
  - Predict the next token `xₜ` given all previous tokens `x₁, x₂, ..., xₜ₋₁`.
  - `P(xₜ | x₁, x₂, ..., xₜ₋₁)` — maximized via cross-entropy loss.
- **GPT Model Progression:**

| Model | Year | Parameters | Key Feature |
|---|---|---|---|
| GPT-1 | 2018 | 117M | First generative pre-trained Transformer |
| GPT-2 | 2019 | 1.5B | Zero-shot task performance |
| GPT-3 | 2020 | 175B | In-context few-shot learning |
| GPT-3.5 / ChatGPT | 2022 | ~175B | RLHF fine-tuned for dialogue |
| GPT-4 | 2023 | Undisclosed | Multimodal, significantly improved reasoning |

- **Causal Masking:** The decoder uses a causal attention mask — token at position `t` can only attend to tokens at positions `1` to `t` (no peeking at future tokens).
- **RLHF (Reinforcement Learning from Human Feedback):** Used to fine-tune GPT into ChatGPT — aligns model outputs with human preferences for helpfulness and safety.
- **BERT vs GPT:**

| Feature | BERT | GPT |
|---|---|---|
| Architecture | Encoder-only | Decoder-only |
| Directionality | Bidirectional | Unidirectional (left-to-right) |
| Pre-training | MLM + NSP | Causal LM |
| Best for | Classification, NER, QA | Text generation, completion |

## Key Takeaways
- GPT is a decoder-only Transformer pre-trained via Causal Language Modeling (next token prediction).
- Causal masking ensures the model only attends to previous tokens — essential for autoregressive generation.
- GPT-3's in-context learning (few-shot prompting) enables task performance without fine-tuning.
- RLHF transforms GPT into instruction-following assistants like ChatGPT.
- BERT excels at understanding tasks; GPT excels at generation tasks — knowing this distinction is critical.

```python
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2 model and tokenizer (TensorFlow version)
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = TFGPT2LMHeadModel.from_pretrained(model_name)
```

```python
import tensorflow as tf

input_text = "Once upon a time in a land far, far away"
input_ids = tokenizer.encode(input_text, return_tensors="tf")

# Generate text
generated_outputs = model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    no_repeat_ngram_size=2,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    temperature=1.0,
    pad_token_id=tokenizer.eos_token_id
)

# Decode the output
generated_text = tokenizer.decode(generated_outputs[0], skip_special_tokens=True)
print(generated_text)
```
