# Attention Mechanism

## Definition
The Attention Mechanism is a technique that allows a neural network to dynamically focus on the most relevant parts of the input when producing each element of the output. Introduced by Bahdanau et al. (2015) to address the Seq2Seq bottleneck problem, it computes a weighted sum of all encoder hidden states for each decoder step, where the weights reflect the relevance of each input position to the current output position.

## Why It Is Needed
- **Bottleneck Solution:** In Seq2Seq models, compressing the entire input into a single vector causes information loss — attention allows the decoder to directly access all encoder states.
- **Dynamic Focus:** Different output tokens depend on different parts of the input — attention learns this alignment automatically.
- **Interpretability:** Attention weights provide a visual alignment map between source and target sequences, making models more interpretable.

## Real-World Applications
- Neural Machine Translation with dynamic source word alignment
- Document summarization with focus on key sentences
- Image captioning with spatial attention on relevant image regions
- Speech recognition with temporal attention on audio frames
- Question answering with attention over passage tokens

## Important Points
- **Additive Attention (Bahdanau):** Computes alignment scores using a feedforward network.
  - `eᵢⱼ = vᵀ · tanh(Wₛ · hₛᵢ + Wₕ · hₜⱼ)`
- **Multiplicative Attention (Luong):** Computes alignment scores using dot product between encoder and decoder states. Simpler and faster.
  - `eᵢⱼ = hₛᵢᵀ · hₜⱼ`
- **Softmax Normalization:** Raw scores are converted to a probability distribution (attention weights):
  - `αᵢⱼ = softmax(eᵢⱼ) = exp(eᵢⱼ) / Σ exp(eₖⱼ)`
- **Context Vector:** Weighted sum of encoder hidden states:
  - `cⱼ = Σ αᵢⱼ · hₛᵢ`
- **Alignment Matrix:** A visualization of attention weights across all source-target token pairs — shows which input tokens are attended to for each output token.

## Key Takeaways
- Attention lets the decoder selectively focus on different encoder states for each output token.
- It solves the fixed context vector bottleneck of vanilla Seq2Seq models.
- Bahdanau (additive) and Luong (multiplicative/dot-product) are the two classic attention variants.
- Attention weights form an alignment matrix that can be visualized for model interpretability.
- The attention mechanism is the conceptual foundation of the Transformer architecture.
