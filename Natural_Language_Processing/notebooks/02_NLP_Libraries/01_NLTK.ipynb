# NLTK (Natural Language Toolkit)

## Definition
NLTK (Natural Language Toolkit) is one of the oldest and most comprehensive Python libraries for NLP. It provides tools for tokenization, POS tagging, parsing, stemming, lemmatization, and access to over 50 corpora and lexical resources including WordNet. It is widely used in academia and education for teaching and researching NLP concepts.

## Why It Is Needed
- **Educational Foundation:** NLTK is the most beginner-friendly NLP library, ideal for learning core NLP concepts step by step.
- **Rich Resources:** Provides access to linguistic databases like WordNet, Penn Treebank, and Brown Corpus out of the box.
- **Comprehensive Toolset:** Covers the full NLP preprocessing pipeline from tokenization to semantic analysis in a single library.

## Real-World Applications
- Academic research and prototyping NLP pipelines
- Teaching NLP fundamentals in courses and workshops
- Linguistic analysis of text corpora
- Building rule-based NLP components for text processing
- WordNet-based synonym and semantic similarity lookups

## Important Points
- **Installation:** `pip install nltk` + `nltk.download('all')` to download corpora and models.
- **Key Modules:**
  - `nltk.tokenize` — Word and sentence tokenizers
  - `nltk.corpus` — Access to 50+ linguistic corpora
  - `nltk.stem` — Stemmers (Porter, Snowball, Lancaster)
  - `nltk.tag` — POS taggers
  - `nltk.chunk` — Chunkers for shallow parsing
  - `nltk.sentiment` — VADER sentiment analyser
- **WordNet:** A large lexical database of English — synsets, hypernyms, hyponyms, antonyms, and semantic similarity.
- **VADER:** A rule-based sentiment analyser in NLTK optimized for social media text.
- NLTK is slower and less production-ready than spaCy but excellent for learning.
