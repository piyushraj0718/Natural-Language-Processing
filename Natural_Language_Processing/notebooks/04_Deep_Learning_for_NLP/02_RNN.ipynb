# RNN (Recurrent Neural Network)

## Definition
A Recurrent Neural Network (RNN) is a class of neural network designed to process sequential data by maintaining a hidden state that is updated at each time step. At every step, the network takes the current input and the previous hidden state to produce a new hidden state and output. This recurrence allows RNNs to model temporal dependencies in sequences like text, speech, and time-series data.

## Why It Is Needed
- **Sequential Modeling:** Standard feedforward networks cannot handle variable-length sequences — RNNs process inputs step by step, maintaining context across the sequence.
- **Context Preservation:** The hidden state acts as a memory that carries information from previous tokens to future ones.
- **Foundation for Advanced Models:** RNNs introduced the core concept of sequential processing that evolved into LSTMs, GRUs, and eventually Transformers.

## Real-World Applications
- Language modelling and text generation
- Sentiment analysis on sequential text
- Speech recognition (audio frame sequences)
- Time-series prediction and anomaly detection
- Part-of-Speech tagging and sequence labeling

## Important Points
- **Recurrence Formula:**
  - `hₜ = tanh(Wₓ · xₜ + Wₕ · hₜ₋₁ + b)`
  - `yₜ = Wᵧ · hₜ + bᵧ`
  - Where `hₜ` = hidden state at time `t`, `xₜ` = input at time `t`.
- **Shared Weights:** The same weight matrices `Wₓ` and `Wₕ` are reused at every time step.
- **Vanishing Gradient Problem:** During backpropagation through time (BPTT), gradients shrink exponentially over long sequences — RNNs struggle to learn long-range dependencies.
- **Exploding Gradient Problem:** The opposite of vanishing — gradients grow uncontrollably, addressed by gradient clipping.
- **BPTT (Backpropagation Through Time):** The algorithm for training RNNs by unrolling the network through time and computing gradients at each step.

## Key Takeaways
- RNNs process sequences step by step, maintaining a hidden state as memory.
- The vanishing gradient problem prevents RNNs from learning long-range dependencies.
- BPTT (Backpropagation Through Time) is the training algorithm for RNNs.
- Gradient clipping is used to prevent the exploding gradient problem.
- LSTMs and GRUs were developed specifically to overcome the limitations of vanilla RNNs.
