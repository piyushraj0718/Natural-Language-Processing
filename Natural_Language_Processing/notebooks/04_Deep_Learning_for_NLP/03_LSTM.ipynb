# LSTM (Long Short-Term Memory)

## Definition
LSTM (Long Short-Term Memory) is an advanced type of Recurrent Neural Network designed to overcome the vanishing gradient problem of vanilla RNNs. Introduced by Hochreiter & Schmidhuber (1997), LSTMs use a gating mechanism with three gates — Forget, Input, and Output — to selectively remember or discard information over long sequences, enabling the learning of long-range dependencies.

## Why It Is Needed
- **Long-Range Dependencies:** Vanilla RNNs fail to retain information across long sequences — LSTMs use gates to control what is remembered and what is forgotten.
- **Vanishing Gradient Solution:** The cell state `Cₜ` provides an uninterrupted gradient flow highway, mitigating the vanishing gradient issue.
- **Selective Memory:** Gates allow the network to learn which information to keep, update, or discard — making it far more powerful than simple RNNs.

## Real-World Applications
- Machine translation (encoder-decoder LSTMs)
- Speech recognition and text-to-speech synthesis
- Sentiment analysis on long documents
- Named Entity Recognition (NER) sequence labeling
- Music generation and handwriting recognition

## Important Points
- **Two States:** LSTM maintains two states:
  - `hₜ` — Hidden State (short-term memory, output)
  - `Cₜ` — Cell State (long-term memory)
- **Three Gates:**
  - **Forget Gate (`fₜ`):** Decides what to erase from cell state. `fₜ = σ(Wf · [hₜ₋₁, xₜ] + bf)`
  - **Input Gate (`iₜ`):** Decides what new information to write to cell state. `iₜ = σ(Wi · [hₜ₋₁, xₜ] + bi)`
  - **Output Gate (`oₜ`):** Decides what to output from the cell state. `oₜ = σ(Wo · [hₜ₋₁, xₜ] + bo)`
- **Cell State Update:** `Cₜ = fₜ ⊙ Cₜ₋₁ + iₜ ⊙ tanh(Wc · [hₜ₋₁, xₜ] + bc)`
- **Hidden State Update:** `hₜ = oₜ ⊙ tanh(Cₜ)`
- `σ` = Sigmoid activation (outputs 0–1, used for gates); `⊙` = element-wise multiplication.

## Key Takeaways
- LSTM solves the vanishing gradient problem through its gated cell state mechanism.
- Three gates — Forget, Input, Output — control the flow of information selectively.
- The cell state `Cₜ` is the long-term memory; the hidden state `hₜ` is the short-term output.
- Sigmoid gates output values between 0 and 1, acting as soft on/off switches.
- LSTMs were the dominant sequence model before Transformers replaced them for most NLP tasks.
